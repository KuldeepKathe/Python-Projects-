```python
import pandas as pd            # impoerting pandas library
import requests                # impoerting requests library.
from bs4 import BeautifulSoup  # impoerting BeautifulSoup library.

Product_name = []  # here we create empty list for Product name,prices,descriptions,rating /reviews.
Prices = []
Description = []
Reviews = []


# for 10 pages data from flipkart website you can just put the all code in for loop.

#for i in range(2,10):  

# url is variable and inside it flipkart website link for mobile under 50,000. This is 1 page mobile Data.

url = "https://www.flipkart.com/search?q=mobile+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(1) 

r = requests.get(url)  # requests for getting info from enter url.
        
soup = BeautifulSoup(r.text,"lxml")

box = soup.find("div",class_ = "_1YokD2 _3Mn1Gg")  # box variable create for only one page and 24 nos. mobile and its reviews.


names = box.find_all("div",class_ = "_4rR01T")   # here we find name of the mobile and iterate in loop for 1 page.
   
for i in names:
    name = i.text
    Product_name.append(name)
#print(len(Product_name))

prices = box.find_all("div",class_ = "_30jeq3 _1_WHN1")  # here we find prices of the mobile and iterate in loop for 1 page.

for i in prices:
    name = i.text
    Prices.append(name)
#print(len(Prices))

desc = box.find_all("ul",class_="_1xgFaf")    # here we find description of the mobile and iterate in loop for 1 page.

for i in desc:
    name = i.text
    Description.append(name)
#print(len(Description))

reviews = box.find_all("div",class_ = "_3LWZlK")  # here we find reviews of the mobile and iterate in loop for 1 page.
    
for i in reviews:
    name = i.text
    Reviews.append(name)
#print(len(Reviews))

df = pd.DataFrame({"Product Name":Product_name,"Prices":Prices,"Description":Description,"Reviews":Reviews}) # dataframe
print(df)

df.to_csv(r"K:\IT (VEDANT)\wap scraping project\WAP_Scrape_Project_Flipcart.csv")  # save data into csv file.
```
